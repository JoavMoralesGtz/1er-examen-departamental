Ejercicio 1

```
print("\tPrograma para calcular una ecuacion de segundo orden.")

x=1

while x!=0:
  a= float (input("\nDame el termino cuadratico: "))
  b= float (input("Dame el termino con variable: "))
  c= float (input("Dame el termino independiente: "))

  r1= 4*a*c
  rn= b**2
  rm= rn - r1
  r2= rm**(1/2)
  r3= -b+r2
  r4= r3/(2*a)
  r5= -b-r2
  r6= r5/(2*a)
  if rm < 0:
    print("\n\t Error. El discriminante debe de ser positivo.")
  else:
    print("\n\nEl primer resultado es: " , r4)
    print("El segundo resultado es: " , r6)
    break



```
